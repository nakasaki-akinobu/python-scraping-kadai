```python
import requests
import pprint
from getpass import getpass
```

```python
# APIキーを指定
api_key = getpass("APIキーを入力してください：")
```

```python
# 検索キーワードを変数station_nameに格納
station_name = input("検索する駅を入力してください：")
```

```python
# APIエンドポイントURL（APIアクセス接続先）
url = 'https://maps.googleapis.com/maps/api/geocode/json'

# データを取得
response = requests.get(
    url,
    params={
        'key': api_key,
        'address': station_name
    }
)
```

```python
# レスポンスのJSONを取得
json_data = response.json()

# JSONの中身を確認
pprint.pprint(json_data)
```

```python
for item in json_data["results"]:
    lat = item['geometry']['location']['lat']
    lng = item['geometry']['location']['lng']
    print(f'{lat}: {lng}')
```

```python
url_restaurant = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

response_restaurant = requests.get(
    url_restaurant,
    params={
        "location": f'{lat},{lng}',
        "radius": 500,
        'language': 'ja',
        "type": 'food',
        "key": api_key,
        'business_status': 'OPERATIONAL',
        'maxResults': 10
    }
)
```

```python
json_restaurant_data = response_restaurant.json()
pprint.pprint(json_restaurant_data)
```

```python
# レストラン名、評価、住所を表示
for item in json_restaurant_data['results']:
    restaurant_name = item['name']
    evaluation = item.get('rating', 'N/A')
    address = item['vicinity']
    print(f'{restaurant_name}:{evaluation}: {address}')
```
